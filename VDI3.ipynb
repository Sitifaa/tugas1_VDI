```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. menentukan data yang akan dilakukan visualisasi data, yaitu data penjualan buku terlaris yang didapatkan pada website kaggle

```python
# Membaca file Excel
data = pd.read_csv('best-selling-books.csv')
print(data.head())
```

```python
data.info()
```

```python
# Menghapus baris dari indeks 31 hingga 174
data = data.drop(data.index[31:175])
print(data.head())
```

a. Tujuan (Objective):
Tujuan dari visualisasi ini adalah untuk memberikan informasi yang jelas dan mudah dipahami terkait dengan data buku terlaris, misalnya penjualan, kategori, tahun penerbitan, atau penulis yang populer.

b. Pengguna (User):
Pengguna visualisasi ini adalah:

Penggemar buku yang tertarik dengan tren buku terlaris.
Penulis atau penerbit yang ingin mengetahui pola dalam penjualan buku.
Peneliti atau analis yang ingin menganalisis data untuk memahami tren pasar buku.

c. Fungsi (Function):
Visualisasi ini harus memungkinkan pengguna:
Melihat tren penjualan buku selama beberapa tahun.
Mengidentifikasi penulis populer atau kategori buku yang paling laris.
Memahami distribusi data seperti kategori buku atau tahun terbit.

d. Tone:
Tone dari visualisasi ini harus bersifat informal namun profesional karena pengguna terdiri dari kelompok yang beragam, termasuk penulis, penggemar buku, dan analis.

3. visualisasinya

a. genre buku yang sering dibeli
terlihat genre fantasy merupakan genre buku yang sering dibeli oleh pembaca

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='Genre', order=data['Genre'].value_counts().index, palette='coolwarm')
plt.title('Genre Buku Terlaris')
plt.xlabel('Jumlah Buku')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='First published', y='Approximate sales in millions', marker='o', color='green')
plt.title('Tren Penjualan Buku Berdasarkan Tahun Pertama Terbit')
plt.xlabel('Tahun Pertama Terbit')
plt.ylabel('Penjualan')
plt.tight_layout()
plt.show()
```

```python
author_sales = data.groupby('Author(s)')['Approximate sales in millions'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=author_sales.values, y=author_sales.index, palette='viridis')
plt.title('Penulis dengan Penjualan Terbanyak')
plt.xlabel('Penjualan')
plt.ylabel('Penulis')
plt.tight_layout()
plt.show()
```
